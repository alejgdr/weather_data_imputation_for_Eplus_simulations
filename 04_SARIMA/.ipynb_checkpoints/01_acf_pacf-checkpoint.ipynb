```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pandas.plotting import register_matplotlib_converters
from statsmodels.tsa.stattools import acf, pacf
import scipy.stats as st
```

```python
dfrad=pd.read_csv('../01_weather_data/02_cleaned_data/TMY_night_solar_angles.csv',index_col=0,parse_dates=True)
dfrad['minutodia']=(dfrad.index.hour*60)+dfrad.index.minute
dfrad
```

```python
dfmingroup=dfrad.groupby(['minutodia',pd.Grouper(freq='1H')]).mean()
dfmindia=dfmingroup.groupby(pd.Grouper(level='minutodia',axis=0)).mean()
# dfmingroup#['Global','prediccion','mae']
```

```python
plt.figure(figsize=(20,10))
data=dfrad.G.loc['2018-01':'2018-1-10'].diff(144)
plt.plot(data,alpha=.5)
plt.plot(data.rolling(144).mean())
plt.plot(data.rolling(144).std())
```

```python
plt.figure(figsize=(20,10))
df_diff=dfmingroup.Ig.loc[600].loc['2018-01':'2018-11'].diff()
plt.plot(df_diff)
plt.plot(df_diff.rolling(30).mean())
plt.plot(df_diff.rolling(30).std())
```

```python
data=dfmingroup.Ig.loc[500].loc['2018-01':'2018-11'][2:100].diff()
# st.t.interval(alpha=0.2, df=len(data[2:100])-1, loc=np.mean(data[2:100]), scale=st.sem(data[2:100])) 
st.norm.interval(alpha=0.90, loc=np.mean(data), scale=st.sem(data))
# df_diff[2:100]
np.mean(data)
```

```python
data=dfmingroup.Ig.loc[680].loc['2018-01':'2018-12'].diff()
# tolerance=.99*(np.mean(data)/np.sqrt(len(data)))
tolerance=(1.96/np.sqrt(len(data[145::])))
tolerance
```

```python
fig, ax=plt.subplots(figsize=(20,10))
# ax.plot(pacf(tmx_2018.Ig,nlags=14))
ax.plot(pacf(data[2::],nlags=30),'r.-')
# plt.bar(range(30),pacf(data[2::],nlags=30)[::30])
ax.axhline(y=tolerance,linestyle='--')
ax.axhline(y=-tolerance,linestyle='--')
ax.axhline(y=0,linestyle='-')
# # ax.plot(pacf(tmx_2018.Ig[5:int(144*7)],nlags=144*3))
# # plt.bar(range(lags),pacf(tmx_2018.Ig[5:int(144*7)],nlags=144*4))
# plt.grid()
# ax.set_xlim(0)
# pacf(df_diff,nlags=144)
```

```python
lags=30
acf_vals=acf(data[2::],nlags=lags)
fig, ax=plt.subplots(figsize=(20,10))
# # plt.figure(figsize=(20,10))
plt.bar(range(lags),acf_vals[:lags])
ax.axhline(y=tolerance,linestyle='--')
ax.axhline(y=-tolerance,linestyle='--')
ax.axhline(y=0,linestyle='-')
# # ax.set_xlim(144*3)
# # ax.set_ylim(-.15,.15)
ax.grid()
# acf_v
```
