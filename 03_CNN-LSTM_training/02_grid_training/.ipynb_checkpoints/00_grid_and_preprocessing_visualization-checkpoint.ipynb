```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from dateutil.parser import parse
from datetime import timedelta
import matplotlib.dates as mdates
```

```python
params={'text.latex.preamble':'\\usepackage{lmodern,amsmath}',
       'text.usetex':True,
       'font.size':10,
       'font.family':'lmodern'}
# h_fmt=mdates.DateFormatter('%m-%d-%H')
plt.rcParams.update(params)
tmx_2019=pd.read_csv('../../01_weather_data/02_cleaned_data/esoru_night_angles.csv',index_col=0,parse_dates=True)
```

```python
hora_input='2019-05-25 9:00:00'
hora_output='2019-05-26 9:00:00'
dia1='2019-05-25'
dia2='2019-05-26'


fig, ax=plt.subplots(figsize=(5.63,2.8))
ax.plot(tmx_2019.Global.loc[hora_input:parse(hora_input)+pd.Timedelta('50min')],
        'g.-',label='Input',linewidth=.9,markersize=3)
ax.plot(tmx_2019.Global.loc[hora_output:parse(hora_output)+pd.Timedelta('50min')],
        'r.-',label='Target',linewidth=.9,markersize=3)
ax.plot(tmx_2019.Global.loc[dia1:parse(dia1)+pd.Timedelta('2D')],
        'k-',alpha=.7,label='$I_g$',linewidth=.9,markersize=3)
ax.legend(fontsize=9.5)
# ax.legend.loc('upper_left')
ax.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d-%H'))
ax.set_ylabel('$I_g [\dfrac{W}{m^2}]$')# [\dfrac{W}{m^2}$]')
ax.set_xlabel('Time [mm-dd-HH]')

ax.grid(alpha=.3)

fig.autofmt_xdate()
plt.tight_layout()
# plt.savefig('../../Tesis_repo/figures/input_target.pdf')
```
