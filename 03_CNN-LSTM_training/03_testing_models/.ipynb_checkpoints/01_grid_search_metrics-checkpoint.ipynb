```python
nombres1=['tiempo','Direct','Global','Difusa','Temperatura','Humedad','Presion','alturasolar','azimuth']#['tiempo','Direct','Global','Difusa','Temperatura','Humedad','Presion','alturasolar','declinacion']
esoru=tst.importa('../../01_Documentos/02_preprocessed/ESORU_noche_angulossolares.csv',nombres1)
tmy=tst.importa('../../01_Documentos/02_preprocessed/TMY_noche_angulossolares.csv',nombres1)
path='../01_Documentos/05_hypopt/'
listmodels=['out_s_144_5_CNN_5_Max_5_LSTM_3_dense_lr_0.0001_in_s_144Global_azimuthaltura.h5'
           ,'out_s_144_5_CNN_5_Max_5_LSTM_3_dense_lr_0.0001_in_s_288Global_azimuthaltura.h5']
# listmodels=['out_s_144_5_CNN_5_Max_5_LSTM_3_dense_lr_0.0001_in_s_144Global_azimuthaltura.h5','out_s_144_5_CNN_5_Max_5_LSTM_3_dense_lr_0.0001_in_s_288Global_azimuthaltura.h5','out_s_144_5_CNN_5_Max_5_LSTM_3_dense_lr_0.0001_in_s_576Global_azimuthaltura.h5']
in_size_list=[144,288,576]
for heo in range(len(listmodels)):
    model5=tst.load_model(path+listmodels[heo])
    in_size,out_size= 576,144 #model4
    #in_size, out_size =72,18 #model3
    dias_rango=360#350#=time_hor
    istep=0#49400-288#6400 
    #x_array,y_array=pre_process(fecha1,fecha2,434,144,esoru)
    scaler3=load(open('../../01_Documentos/03_Models/y_scalerv04.pkl','rb'))
    scaler4=load(open('../../01_Documentos/03_Models/x_scalerv04.pkl','rb'))
    fecha1='01/01/2019'#'01/10/2019'
    fecha2='31/12/2019'#'31/12/2019'
    rango=out_size*int((dias_rango*144)/out_size)
    inputs=['Global','Direct','Temperatura','Humedad','azimuth','alturasolar']
    predi4,targ4=tst.Multioneshot(fecha1,fecha2,esoru,rango,out_size,in_size,istep,model5,inputs,scaler2=scaler3,scaler=scaler4,prep=1)
    mse,rmse,mae=tst.metrics(predi4,targ4,rango)
    nombres=['tiempo','Direct','Global','Difusa','Temperatura','Humedad','Presion','alturasolar','azimuth']
    impesoru=tst.exporta('../../01_Documentos/02_preprocessed/ESORU_noche_angulossolares.csv',predi4,istep,in_size,rango,nombres,sol_data_correction=True)
    impesoru_target=tst.importa('../../01_Documentos/02_preprocessed/ESORU_noche_angulossolares.csv',nombres)
    dfrad=impesoru_target.copy()
    dfrad['prediccion']=impesoru.Global
    dfrad['minutodia']=(dfrad.index.hour*60)+dfrad.index.minute
    dfrad['mse']=(dfrad.Global-dfrad.prediccion)**2
    dfrad['mae']=np.abs(dfrad.Global-dfrad.prediccion)
    dfmingroup=dfrad.groupby(['minutodia',pd.Grouper(freq='1H')]).mean()
    dfmindia=dfmingroup.groupby(pd.Grouper(level='minutodia',axis=0)).mean()
    dfmindia.loc[dfmindia.alturasolar<0,'mae']=np.nan
    meandiay=dfmindia.mae.mean()
    dfsamp4=dfrad.resample('D').sum()
    dfsamp3=dfrad.resample('D').mean()
    #dfsamp3['rmse']=dfsamp3['mse']**.5
    dfsamp3['dif_energia_wh']=dfsamp4['mae']/6
    dfsamp3['energia_wh']=dfsamp4['Global']/6
    dfsamp3['porcentaje_error']=(dfsamp3['dif_energia_wh']/dfsamp3['energia_wh'])*100
    #dfsamp3=pd.concat(dfsamp,dfsamp.energia)
    tablita=['dif_energia_wh','porcentaje_error','mae_de_d√≠a_promedio']
    #dfsamp3.resample('M').mean()
    dfsamp5=dfsamp3.resample('Y').mean()
    dfsamp5['mae_de_d√≠a_promedio']=meandiay
    print('in_size:',in_size)
    print('model:',listmodels[heo])
    print(dfsamp5[tablita])
```
